```python
#カテゴリ変数をone hot encoding
cat_columns = ['company', 'topping', 'size']
df_cat = data[cat_columns]
cat_data = pd.get_dummies(df_cat)
cat_data.head()
```

```python
data = pd.concat([data, cat_data], axis=1).drop(cat_columns, axis=1)
```

```python
#いらないカラムを削除
drop_columns = ['variant', 'extra_sauce', 'extra_cheese']
data = data.drop(drop_columns, axis=1)
data.head()
```

```python
#price_rupiahを数値に
data['price_rupiah'] = data['price_rupiah'].str.strip('Rp')

target = ','
for index, price in enumerate(data['price_rupiah']):
    idx = price.find(target)
    before = price[:idx]
    after = price[idx+1:]
    price = before + after
    data['price_rupiah'][index] = price
    
data['price_rupiah'] = data['price_rupiah'].astype(int)
```

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression

X_train = data.drop('price_rupiah', axis=1).copy()
y_train = data['price_rupiah']

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```
